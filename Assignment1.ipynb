# Create a plot comparing kohli, Rohit Sharma, and Shewag across 10 years of hypothetica; runs
use:
1. labels
2. legends
3. colors
4. line style
5. one custom style

```python
import matplotlib.pyplot as plt
```

```python
years = [2000,2001,2002,2003,2004,2005,2006,2007,2008,2009]
kohli = [200,400,600,800,1000,1200,1400,1600,1800,1900]
shewag = [400,800,1200,1800,2400,3000,3400,3800,4000,4400]
rohit = [600,1200,1800,2400,3000,3600,4200,4800,5400,6000]

plt.plot(years, kohli, color = "red", linestyle = '--', label = "Virat Kohli")
plt.plot(years, shewag, color = "cyan", linestyle = ':', label = "Virendra Shewag")
plt.plot(years, rohit, color = "blue", linestyle = '-.', label = "Rohit Sharma")

plt.legend()
plt.tight_layout()
plt.style.use('classic')
plt.show()
```
