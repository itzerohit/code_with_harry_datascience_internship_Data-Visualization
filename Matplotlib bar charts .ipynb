# Matplotlib Bar Charts 
* Bar charts are used to compare quantities across categories. They are easy to read and powerful for visual analysis.

# Sachine's Yearly runs - vertical Bar charts '

Basic Bar Charts 

```python
import matplotlib.pyplot as plt 
years = [1990,1992,1994,1996,1998,2000,2002,2004,2006,2008]
runs = [1000,2000,3000,4000,5000,6000,7000,8000,9000,10000]

plt.bar(years,runs)
plt.xlabel("Years")
plt.ylabel("Runs")
plt.title("Sachine Yearly's Performance")
plt.legend()
plt.tight_layout()
plt.show()
```

# Setting Bar width and side-by side Bar Charts 
* The value of width ranges from 0 to 1 (default 0.8); it can go above 1,  but bars will start to overlap

```python
import numpy as np 
import matplotlib.pyplot as plt
years = [2000,2002,2004,2006,2008,2010]
Kohli = [300,400,500,600,700,800]
Sachin = [1000,2000,3000,4000,5000,6000]
Shewag = [600,800,1400,2000,2600,3200]
x = np.arange(len(years))
width = 0.1
plt.bar(x-width, Kohli, color = "red", label = "Virat Kohli")
plt.bar(x, Sachin, color = "blue", label = "Sachin Tendulkar")
plt.bar(x+width, Shewag, color = "cyan", label = "Virendra Shewag")
plt.legend()
plt.tight_layout()
plt.xticks(x,years)
plt.show()
```

# Horizontal Bar CHart 

```python
# Let's compare total runs scored in the first 5 yearts by different players 
players = ["Sachin", "Yuvraj", "Ganguly", "Rahul", "Shewag"]
Runs = [100+200+300+400+500, +700+800+900+1000+1100,  +1300+1400+1500+1600+1700, +1900+2000+2100+2200+2300, 2400+2500+2600+2700+2800]
# Plotting with barh 
plt.barh(players,Runs, color = "red")
plt.legend()
plt.tight_layout()
plt.xlabel("Runs Scored")
plt.ylabel("Players")
plt.title("Players 5-Year performace of Indian-Batsman")
plt.show()
```

# One of the most important things to take 
* In bar() -- x = categories, y = values
* In barh() -- y = categories, x = values 
* Horizontal bars help when category names are long or when you want to emphasize comparisons from left to right 

# Adding valu with plt.text()"

```python
import matplotlib.pyplot as plt
players = ["Sachin","Rohit","Rahul"]
runs = [20000,15000,14000]

plt.bar(players,runs,color="green")
# If we want to add labels at top 
for i in range(len(players)):
    plt.text(i,runs[i]+50, str(runs[i]), ha="center")
plt.ylabel("Runs")
plt.xlabel("Players")
plt.title("Player Runs")
plt.legend()
plt.tight_layout()
plt.show()
```

# Summary of the  barchart 
* plt.bar() -- vertical bars for categorical comparison 
* plt.barh() -- Horizontal bars (great for long labels)
* width = 0.25 -- Control thickness/spacing of bars 
* np.arange() --  Helps to align multiple bars side by side 
* plt.xticks() -- Replace index label with real numbers 
* plt.tight_layout() -- prevent label from overlapping
