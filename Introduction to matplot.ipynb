# Multiple lines in one plot

```python
years = [2000,2001,2002,2003,2004,2005,2006,2007,2008,2009]
kohli = [0, 0, 500, 800, 1100, 1300, 1500, 1800, 1900, 2100]
sehwag = [0, 300, 800, 1200, 1500, 1700, 1600, 1400, 1000, 0]

plt.plot(years, kohli, label="Virat Kohli")
plt.plot(years, sehwag, label="Virender Sehwag")

plt.xlabel("Year")
plt.ylabel("Runs Scored")
plt.title("Performance Comparison")
plt.legend()
plt.show()
```

# Line Width and Layout Tweaks

```python
plt.plot(years,kohli, linewidth = 3, label = "Virat Kohli")
plt.plot(years,shewag, linewidth = 4, label = "Virendra shewag")
plt.plot(years, sachin, linewidth = 5, label = "Sachin Tendulkar")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

XKCD Comic Style

```python
with plt.xkcd():
    plt.plot(years, kohli, label = "Virat Kohli")
    plt.plot(years, shewag, label = "Virendra Shewag")
    plt.plot(years, sachin, label = "Sachin Tendulkar")
    plt.legend()
    plt.xlabel("Years")
    plt.ylabel("Runs")
    plt.title("Batsman Performace Comparison")
    plt.show()
```

# Visualizing tons of data 

```python
import numpy as np
for i in range(10):
    plt.plot(np.random.rand(10), linewidth = 1)

plt.title("Number Comparison")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

# Final Tip For begineers 
1. Always starts with simple plots 
2. Add labels and legends early
3. Use plt.grid(True) or plt.tight_layout() to improve readability
4. Try diffrent styles to find what works for your use case
