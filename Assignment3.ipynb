Create a pie chart showing the market share of mobile OS (Android, iOS, others). Then recreate the same using a horizontal bar chart and observe which is easier to understnad 

```python
import matplotlib.pyplot as plt 
mobile = ["Android", "ios", "nokia", "micromax"]
share = [30,40,20,10]
plt.pie(share,labels=mobile,autopct='%1.1f%%',startangle=140)
plt.title("Market Share of mobile OS")
plt.show()
```

# Now we have to this same questionw with the help of bar chart 

```python
import matplotlib.pyplot as plt 
mobile = ["Android", "ios", "nokia", "micromax"]
share = [30,40,20,10]
plt.barh(mobile,share)
plt.legend()
plt.tight_layout()
plt.xlabel("Number of Share Holding")
plt.ylabel("Name of the OS Company")
plt.title("Market Share of mobile OS")
plt.show()
```
