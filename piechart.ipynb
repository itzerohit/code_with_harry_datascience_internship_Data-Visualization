# Matplotlib Pie charts 
* pie charts are used to show part-to-whole relationships. They're visually appealing but best used with fewer categires, as too many slices can get cluttered and hard to interpret 

```python
import matplotlib.pyplot as plt
```

```python
print(plt.style.available)
```

```python
plt.style.use('seaborn-v0_8-paper')
labels = ['Dhawan', 'Rohit', 'Kohli','Sachin']
Runs = [10000,20000,30000,40000]
plt.pie(Runs,labels= labels)
```

```python
# like what if i want to change the color based on the na,e 
colors = ['red','blue','pink','blue'] # Even we can add hex code of the color behalf of the color name 
plt.pie(Runs,labels = labels,colors = colors)
plt.title("Total Run's")
```

# ADD edges and styles slices with wedgeprops

```python
plt.pie(Runs,labels=labels,colors=colors,wedgeprops = {'edgecolor':'blue','linewidth':2,'linestyle': '--', 'fill':  False})
```

```python
plt.pie(Runs,labels=labels,colors=colors,wedgeprops = {'edgecolor':'blue','linewidth':2,'linestyle': '--', 'fill':  True})
```

# Highlight a Slice Using explode

```python
# Use explode to pull out one or more slices 
explode = [0,0,0,0.4] # only highlighting rohit sharma 
plt.pie(Runs,labels=labels, explode=explode, colors = colors)
plt.title("Exploded Pie Example")
plt.show()
```

# Add Shadows for a 3D Feel

```python
# Adds a 3D-like shadow
explode = [0,0,0.3,0] # only highlighting rohit sharma 
plt.pie(Runs,labels=labels, explode=explode, colors = colors, shadow = True)
plt.title("Exploded Pie Example")
plt.show()
```

# Startangel 
* Use startangle to rotate the pie for better alignment

```python
# Adds a 3D-like shadow
explode = [0,0,0.3,0] # only highlighting rohit sharma 
plt.pie(Runs,labels=labels, explode=explode, colors = colors, shadow = True, startangle=150, )
plt.title("Exploded Pie Example")
plt.show()
```

```python
# Show Percentages with autopct
explode = [0,0,0,0.2]
plt.pie(Runs, labels=labels, colors=colors, shadow=True, startangle = 230,explode = explode, autopct='%1.1f%%')
plt.title("Career Run Share")
plt.show()
```

# Crowded Pie Chart  -- Why to Avoid 

```python
# Too many categories 
languages = ['Python', 'Java', 'C++', 'JavaScript', 'C#', 'Ruby', 'Go', 'Rust', 'Swift', 'PHP']
usage = [30,20,10,10,7,5,4,3,2,1]
plt.pie(usage,labels = languages, autopct = '%1.1f%%', startangle=90)
plt.title("Langague Usage Comparison")
plt.show()
```

# Why avoid it 
* Hard to compare slice sizes visually 
* Cluttered and confusing 
* No clear insight 
* Better alternative : bar charts or horizontal bar charts 

# Summary of useful Pie charts parameters 
* labels - labels each slice 
* colors - customize slice colors (names or hex)
* explode -- put out slices for emphasis 
* shadow -- Adds depth-like shadow 
* startangle -- Rotates pie to start  from a diffrent angle
* autopct -- Shows percentage text on slices
* Wedgeprops - custommize slice edge, fill, width, etc.
