# Try making a stackplot with your own weekly schedule

```python
import matplotlib.pyplot as plt 

days = [1,2,3,4,5,6,7]
gym = [1,2,1,2,1,3,2]
breakfast = [1,2,3,4,3,2,1]
college = [2,3,4,5,4,3,2]
lunch = [1,2,3,1,2,3,1]
study = [3,3,3,3,3,3,3]
youtube = [2,2,2,2,2,2,2]
sleep = [6,7,8,7,6,7,8]


plt.figure(figsize=(10,6))
labels = ["Gym", "Breakfast", "College", "Lunch", "Study", "Youtube", "Sleep"]
colors = ["red", "cyan", "blue", "green", "yellow", "pink", "black"]

plt.stackplot(days, gym, breakfast, college, lunch, study, youtube, sleep, labels = labels , colors = colors, alpha = 0.8)
plt.title("Weekly Scheduled Routine of Mine")
plt.xlabel("Days")
plt.ylabel("Hours")
plt.grid(True)
plt.tight_layout()
plt.show()
```
