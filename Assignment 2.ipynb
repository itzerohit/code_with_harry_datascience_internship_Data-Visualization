* Create a side-by-side bar chart comparing runs of sachin, kohli, and shewag over 5 years. Then create a horizontal bar chart showing total runs scored in their debut 5 years 

```python
import numpy as np 
import matplotlib.pyplot as plt 
years = [2021,2022,2023,2024,2025]
kohli = [1000,2000,3000,4000,5000]
shewag = [6000,7000,8000,9000,10000]
sachin = [11000,12000,13000,14000,15000]
x = np.arange(len(years))
width = 0.25
plt.bar(x-width, kohli, color = "green", width = width, label = "Virat Kohli")
plt.bar(x, shewag, color = "blue", width = width, label = "Virendra Shewag")
plt.bar(x+width, sachin, color = "yellow", width = width, label = "Sachin Tendulkar")  
plt.legend()
plt.tight_layout()
plt.xlabel("Years")
plt.ylabel("Run Scored")
plt.title("Run's Comparison Of Indian Batsman")
plt.grid(True)
plt.show()
```
